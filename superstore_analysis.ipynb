```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/content/sample_data/Sample - Superstore.xls', sheet_name='Orders')
```

```python
# найдем худший заказ
worst_order = df[df['Profit'] == df['Profit'].min()]

worst_order
```

```python
unprofitable_df = df[df['Profit'] < 0]
unprofitable_df.head()
```

```python
# найдем самые проблемные убыточные категории
worst_subcategories = unprofitable_df.groupby('Sub-Category')['Profit'].sum().sort_values()

print(worst_subcategories)
```

Выходит, источники убытков компании — это не случайные провалы, а системные проблемы в трех категориях: "Binders", "Tables" и "Machines".

```python
# сделаем график для наглядности
worst_subcategories.plot(
    kind='bar',
    figsize=(10, 8),
    color='salmon',
    title='Суммарные убытки по подкатегориям',
    xlabel='Суммарный убыток в $',
    ylabel='Подкатегория'
)

# добавим сетку для уддобства
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

Теперь подготовим данные для Tableu.

```python
columns_for_tableau = [
  'Order ID', 'Order Date', 'Customer Name', 'Segment',
    'Country', 'City', 'State', 'Region',
    'Category', 'Sub-Category', 'Product Name',
    'Sales', 'Quantity', 'Discount', 'Profit'
 ]

# создаем финальный DataFrame для выгрузки
tableau_df = unprofitable_df[columns_for_tableau]

output_filename = 'unprofitable_orders_for_tableau.csv'

tableau_df.to_csv(output_filename, index=False)

print(f"Данные успешно сохранены в файл: {output_filename}")
```
